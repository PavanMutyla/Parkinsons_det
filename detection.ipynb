```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as pt
import seaborn as sns

#evaluating
from sklearn.metrics import f1_score,confusion_matrix,matthews_corrcoef,accuracy_score,precision_score
from sklearn.metrics import auc,roc_curve,r2_score,RocCurveDisplay,recall_score

#training,split,pipeline,models
from sklearn.model_selection import train_test_split,GridSearchCV,RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

#using neighbors
from sklearn.neighbors import KNeighborsClassifier
```

## MDVP:Fo(Hz) - Average vocal fundamental frequency
MDVP:Fhi(Hz) - Maximum vocal fundamental frequency
    
MDVP:Flo(Hz) - Minimum vocal fundamental frequency

MDVP:Jitter(%),MDVP:Jitter(Abs),MDVP:RAP,MDVP:PPQ,Jitter:DDP - Several
measures of variation in fundamental frequency

MDVP:Shimmer,MDVP:Shimmer(dB),Shimmer:APQ3,Shimmer:APQ5,MDVP:APQ,Shimmer:DDA - Several measures of variation in amplitude


NHR,HNR - Two measures of ratio of noise to tonal components in the voice

class - Health status of the subject (one) - Parkinson's, (zero) - healthy

RPDE,D2 - Two nonlinear dynamical complexity measures

DFA - Signal fractal scaling exponent

spread1,spread2,PPE - Three nonlinear measures of fundamental frequency variation



```python
data = pd.read_csv("Parkinson disease.csv")
```

```python
data = data.drop(labels=['name'],axis=1)
```

```python
data = data.rename(columns={'status':'class'})
```

```python
X = data.drop('class',axis=1)
```

```python
Y = data['class']
```

```python
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3)
```

```python
## Logistic
clf = LogisticRegression()
```

```python
clf.fit(X_train,Y_train)
```

```python
clf.score(X_test,Y_test)
```

```python
def eval(y_true,y_pred):
    mcc = matthews_corrcoef(y_true,y_pred)
    recall = recall_score(y_true,y_pred)
    f1 = f1_score(y_true,y_pred)
    acc = accuracy_score(y_true,y_pred)
    pre = precision_score(y_true,y_pred)
    r2 = r2_score(y_true,y_pred)
    
    result = {'mcc':mcc,'recall':recall,'f1':f1,'acc':acc,'pre':pre,'r2':r2}
    return result
    
```

```python
logistic_pred = clf.predict(X_test)
```

```python
log_score=eval(Y_test,logistic_pred)
```

```python
logistic_pred
```

```python
#Randomforest
clf1 = RandomForestClassifier()
clf1.fit(X_train,Y_train)
```

```python
rand_preds = clf1.predict(X_test)
```

```python
rand_preds
```

```python
rand_score = eval(Y_test,rand_preds)
```

```python
## KNN
clf2 = KNeighborsClassifier()
```

```python
clf2.fit(X_train,Y_train)
```

```python
KNN_pred = clf2.predict(X_test)
```

```python
KNN_score = eval(Y_test,KNN_pred)
```

```python
before_tuning = {'Logistic':log_score,'Random':rand_score,'KNN':KNN_score}
```

```python
before = pd.DataFrame(before_tuning)
```

```python
be = before.T
```

```python

be.plot(kind='bar')
```

## TUNING 

```python
log_grid= {
    'penalty': ['l1', 'l2'],
    'C': [0.1, 1, 10],
    'solver': ['liblinear'],
}
```

```python
rf_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10],
}
```

```python
KNN_grid={  'n_neighbors': [3, 5, 7]}
```

```python
grid_log = GridSearchCV(LogisticRegression(),param_grid=log_grid,cv=5,n_jobs=5,verbose=True)
```

```python
grid_log.fit(X_train,Y_train)
```

```python
grid_rf = GridSearchCV(RandomForestClassifier(),param_grid=rf_grid,cv=5,n_jobs=5,verbose=True)
```

```python
grid_rf.fit(X_train,Y_train)
```

```python
grid_KNN = GridSearchCV(KNeighborsClassifier(),param_grid=KNN_grid,cv=5,n_jobs=5,verbose=True)
```

```python
grid_KNN.fit(X_train,Y_train)
```

```python
grid_log_pred = grid_log.predict(X_test)
```

```python
grid_rf_pred = grid_rf.predict(X_test)
```

```python
grid_KNN_pred = grid_KNN.predict(X_test)
```

```python
grid_KNN_score=eval(Y_test,grid_KNN_pred)
grid_rf_score=eval(Y_test,grid_rf_pred)
grid_log_score=eval(Y_test,grid_log_pred)
```

```python
after = {"KNN":grid_KNN_score,"LOg":grid_log_score,"rf":grid_rf_score}
```

```python
after_grid = pd.DataFrame(after).T
```

```python
after_grid.plot(kind='bar')
```
